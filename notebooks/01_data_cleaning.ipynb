```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_excel("../data/raw/Data.xlsx")
df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

```python
if "blank" in df.columns:
    df = df.drop(columns=["blank"])

df.head()
```

```python

df.columns
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["ship_date"] = pd.to_datetime(df["ship_date"], errors="coerce")
df[["order_date", "ship_date"]].dtypes

numeric_cols = [
    "sales",
    "profit",
    "discount",
    "shipping_cost",
    "unit_price",
    "order_quantity",
    "product_base_margin",
    "profit_ratio",
    "days_to_ship"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

df[numeric_cols].describe()
```

```python
df[numeric_cols].isnull().sum()
```

### Data Type Validation

- Converted order and shipping date columns to datetime format
- Ensured all sales, profit, discount, cost, and quantity fields are numeric
- Verified that no missing values were introduced during type coercion

```python
df.to_csv("../data/processed/cleaned_sales.csv", index=False)
```

## Data Cleaning Output

The dataset has been cleaned and standardized for downstream analysis.

- Removed irrelevant and empty columns
- Converted order and shipping dates to datetime format
- Ensured numeric consistency for sales, profit, discount, and cost-related fields
